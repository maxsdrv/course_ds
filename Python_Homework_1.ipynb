**Условие.** Напишите программу, которая просит пользователя ввести электронную почту и проверяет корректность её введения. По условию, почтовый адрес является корректным, если в нём присутствуют один и только один символ `@` (при этом слева от этого символа должен находиться хотя бы ещё один [исключая `@`]). _Это не полный список критериев, но для учебного примера достаточно этих ограничений._ При нарушении хотя бы одного условия программа должна ответить, что такая почта не принимается, и разъяснить причину. Если почта корректна, об этом также надо сообщить пользователю.

**Алгоритм решения:**
1. Получить на вход почту в виде строки (функция `input`).
2. Проверить разные условия (условная конструкция и подходящие операторы):
 - проверить наличие ровно одного символа `@` (строковый метод `.count()`) и вывести отказ с пояснением, если проверка не пройдена (функция `print()`);
 - проверить наличие символов перед `@` (строковый метод `.startswith()` — если начинается с `@`, то условие не выполнено) и вывести отказ с пояснением, если проверка не пройдена (функция `print()`).
3. Сообщить пользователю, что почта корректна, если это так (функция `print()`).

**Решение:**

````python
email = input()  # шаг 1

if email.count('@') != 1:  # шаг 2-1
    print('Введённый почтовый адрес некорректен. Он должен содержать ровно одну @')
elif email.startswith('@'):  # шаг 2-2
    print('Введённый почтовый адрес некорректен. Он должен иметь хотя бы один символ перед @')
else:  # шаг 3
    print('Поздравляем, ваша почта принята!')
````

### Задача 1 (5 баллов)

Пользователь подаёт на вход программе действительные числа через пробел. Он привык пользоваться Excel, поэтому разделяет числа запятой, а не точкой. Кроме того, он может случайно ввести число «словами» (например, не "5", а «пять» — в этом случае слово надо игнорировать). **Ваша задача** — вывести минимум и максимум последовательности через запятую с пробелом. Гарантируется, что будет введено хотя бы одно число.

**Пример:**

```
input:
-2.56 25,002 восемь

output:
-2.56, 25.002
```

**Критерии:**

- задача решена корректно (учтено оба условия — разделение чисел запятыми и ввод чисел словами), описание присутствует и соответствует решению — 5 баллов;
- задача решена корректно (учтено оба условия — разделение чисел запятыми и ввод чисел словами), описание отсутствует / задача решена удовлетворительно (учтено одно из условий), описание присутствует и соответствует решению — 4 балла;
- задача решена с серьёзными ошибками, но описание присутствует и соответствует решению — 3 балла;
- иное — 0 баллов.

**Алгоритм решения:**

1. Использую функцию `split()` для разделения строки на отдельные элементы по символу пробела, а затем пройти по каждому элементу и проверить, является ли он числом. Если элемент является числом, то можно добавить его в список чисел.

2. Для проверки, является ли элемент числом, можно использовать функцию `isdigit()` для каждого символа элемента, и если все символы являются цифрами или символом запятой, то элемент можно преобразовать в число. Если элемент не является числом, то можно пропустить его и перейти к следующему элементу.

3. Когда список чисел будет заполнен, можно найти минимум и максимум с помощью функций `min()` и `max()` соответственно.

**Пример решения:**


````python
# ваш код здесь
# Получаем входные данные от пользователя
input_str = input("Введите числа через пробел: ")

# Разбиваем строку на отдельные элементы
input_list = input_str.split()

# Создаем список для чисел
numbers = []

# Проходим по каждому элементу и проверяем, является ли он числом
for item in input_list:
    # Убираем возможные запятые
    item = item.replace(",", "")
    # Проверяем, является ли элемент числом
    if item.isdigit():
        # Преобразуем элемент в число и добавляем его в список
        numbers.append(float(item))

# Находим минимум и максимум списка чисел
minimum = min(numbers)
maximum = max(numbers)

# Выводим результаты
print("Минимум: {}, максимум: {}".format(minimum, maximum))
````

**Критерии:**

**Алгоритм решения:**

1. Для решения этой задачи сначала необходимо распарсить входящую строку на необходимые аргументы для вызовы функции, которую реализуем в дальнейшем.

2. После того как мы успешно распарсили данные мы можем вызыватььнеобходимые нам функции округления, алгоритм функций описан в коде. Могу добавить что он похож на тот алгоритм который под "капотом"      реализован в функциях модуля `math` это `math.ceil()` и `math.floor()`
3. Смотри комментраии в коде.

**Пример решения:**


````python
# ваш код здесь

def parser(a_str): # Функция парсинга входящих данных
    res_str = "" # Объявляем строку 

    it_begin = a_str.find("(") # Находим итератор открывающейся скобки 

    it_end = a_str.find(")")  # Находим итератор закрывающейся скобки

    if it_begin==-1 or it_end==-1: # Если открывающей и закрывающей скобки нет то входящие данные не корректны, выбрасываем исключение
        print("Error, function is not correct\n"
              "Please visit Excel https://...")

    res_str = a_str[it_begin+1:it_end] # Испоьзуем срезы чтобы извлечь число из скобок

    arg_list = res_str.split(";")  # Получаем два числа разделенные пробелом

    arg_list[0] = arg_list[0].replace(",", ".") # Заменяем запятую на точку для корректного преобразования во float

    func_name = a_str[:a_str.find("(")] # С помощью среза и find() находим имя для вызова функции

    number = eval(arg_list[0]) # Преобразуем с помощью eval() строку в число (это превый аргумент для наших функции)
    digit = eval(arg_list[1]) #  Преобразуем с помощью eval() строку в число (это второй аргумент для наших функции)


    # Далее с помощью if-elif-else вызываем необходимую функцию для округления нашего числа
    if func_name == "ОКРУГЛ":
        result = round_rel(number, digit)
    elif func_name == "ОКРУГЛВНИЗ":
        result = round_down(number, digit)
    elif func_name == "ОКРУГЛВВЕРХ":
        result = round_up(number, digit)
    else:
       raise ValueError("Name function doesn't exist")

    return result #


#Эквивалентно ROUND in Excel
def round_rel(number, num_digits):
    # Здесь round() работает так как нам надо, поэтому ничего не трогаем,а просто используем его
    factor = 10 ** num_digits 
    rounded = round(number * factor) 

    return rounded / factor # возвращаем результат

# Эквивалентно ROUNDUP in Excel
def round_up(number, num_digits):
    # Умножаем число на коэффициент 10, 
    # возведенный в степень количества цифр, 
    # до которого мы хотим округлить
    # используем оператор // для выполнения целочисленного деления и округления до ближайшего целого числа,
    # а затем делим округленный результат на тот же фактор
    factor = 10 ** num_digits

    # Для корректного округления в большую сторону при отрицательном числе, нужно для начала убрать знак "-"
    # А потом вернуть его обратно
    s = str(number) 
    if s.startswith("-"): 
        number = (-number) 
        rounded = -(-number * factor // 1) # Для округления в большую сторону
        return (-rounded) / factor  

    # Если число не отрицательное операций по отбрасыванию знака не требуется и мы просто округляем в большую сторону
    rounded = -(-number * factor // 1) # 

    return rounded / factor #

# Эквивалентно ROUNDDOWN in Excel
def round_down(number, num_digits):
    # Алгоритм работы тот же что и round_up, за исключением того что округляем мы в меньшую сторону
    factor = 10 ** num_digits 

    s = str(number)  
    if s.startswith("-"): 
        rounded = -number * factor // 1 # добавляем знак для округления в меньшую сторону
        return (-rounded) / factor 

    else:
        rounded = number * factor // 1 # Здесь знак не нужен 
        return rounded / factor 

def TEST(l1, l2): # Функция для тестов 
    count_err = 0
    count_tests = 0
    for index, req in enumerate(l1):
        if parser(req) != eval(l2[index]):
            count_err += 1
            print(parser(req))
            print(l2[index])
        else:
            count_tests += 1

    if count_err == 0:
        print("ALL ", count_tests, "TESTS SUCCESSFULLY")
    else:
        print("TEST FINISHED with", count_err, "Errors")

###################################################################
###################################################################

if __name__ == '__main__':
    print("HW 2")


#TESTS:

# Round down
test_lst_down = [
    "ОКРУГЛВНИЗ(3,2; 0)",
    "ОКРУГЛВНИЗ(76,9;0)",
    "ОКРУГЛВНИЗ(3,14159; 3)",
    "ОКРУГЛВНИЗ(-3,14159; 1)",
    "ОКРУГЛВНИЗ(31415,92654; -2)"
]

answ_lst_down = [
    "3.0",
    "76.0",
    "3.141",
    "-3.1",
    "31400.0"    
]
print("Start Test excel_ROUNDDOWN")
TEST(test_lst_down, answ_lst_down)

#Round up
test_lst_up = [
    "ОКРУГЛВВЕРХ(3,2;0)",
    "ОКРУГЛВВЕРХ(76,9;0)",
    "ОКРУГЛВВЕРХ(3,14159; 3)",
    "ОКРУГЛВВЕРХ(-3,14159; 1)",
    "ОКРУГЛВВЕРХ(31415,92654; -2)",
]
answ_lst_up = [
    "4",
    "77",
    "3.142",
    "-3.2",
    "31500.0"
]
print("Start Test excel_ROUNDUP")
TEST(test_lst_up, answ_lst_up)


#Round real
test_lst_real = [
    "ОКРУГЛ(2,15; 1)",
    "ОКРУГЛ(2,149; 1)",
	"ОКРУГЛ(-1,475; 2)",
    "ОКРУГЛ(21.5; -1)",
    "ОКРУГЛ(626,3;-3)",
    "ОКРУГЛ(1,98;-1)",
    "ОКРУГЛ(-50,55;-2)",
]
answ_lst_real = [
    "2.2",
    "2.1",
    "-1.48",
    "20",
    "1000",
    "0",
    "-100"
]

print("Start Test excel_ROUND")
TEST(test_lst_real, answ_lst_real)




````

#### Учимся мыслить "питонически"
Чтобы лучше понимать, как устроен и как работает Python, надо научиться воспринимать все в виде классов и объектов. Это далеко не так сложно, как кажется на первый взгляд. К примеру, есть вид _homo sapiens_. Мы можем перечислить его базовые характеристики: двуногий, говорящий, социальный и т. д. Однако в деталях люди, как мы знаем, разные: они отличаются по полу, возрасту, именам, характеру и т. п. И в то же время мы понимаем, что у нас есть очень много общего: мы принадлежим к одному *классу*, являясь каждый уникальным *объектом* этого класса. То есть: класс - идеальная форма, а объект - конкретное выражение этой формы.

В общей сложности, **каждый объект какого-то класса обладает:**

* описательными чертами (что это: какого цвета, сколько весит) - в Питоне мы будем называть это *атрибутами*;
* определенным набором возможных действий (что оно может делать) - в парадигме Python мы будем называть это *методами*.

Далее, нам следует понимать, что Python (как язык программирования) является своеобразной "реальностью" полностью и с нуля сконструированной людьми. Как и в реальном мире, в Питоне происходят какие-то взаимодействия между разными объектами. Разница лишь в том, что в жизни мы изучаем, как это происходит, а в Python мы сами устанавливаем правила взаимодействия. Например, Питон *знает*, что `1 + 1 = 2` по следующим причинам:

* программисты, написавшие Python, определили, что такое целое число (в этом случае `1` - это конкретный объект класса "целове число");
* они же определили (запрограммировали), что будет, если одно целое число встретит `+` в компании другого целого числа.

Вы можете определить СВОЙ СОБСТВЕННЫЙ КЛАСС и самостоятельно решить, как он будет работать при сложении с чем-либо (к примеру, может просто печататься сообщение "Вы прекрасны!").

Так что не следует пугаться, если какой-либо оператор (`+`, `-`, `=` и т. д.) или функция ведут себя неожиданным образом: просто откройте документацию и посмотрите как оно должно работать.

#### Функции
В Питоне вы всегда и везде будете встречать функции. Но что они такое? Что вы помните о них со школы?

Функция, это инструмент, который:

* берет что-то на вход (к примеру, ингридиенты для торта). Мы будем называть это *аргументами функции*;
* выполняет какую-то операцию (например, смешивает ингридиенты вместе);
* что-то возвращает обратно (в нашем случае вкусный торт).

**Важно.** Если вы хотите понять, как работает та или иная функция, вы должны проверить все три пункта. В документации они обычно указываются отдельно.

Обычная функция в Python выглядит так: `function(arg1, arg2, ...)`.

Пример с функцией `print`:

````python
print(1, 2, 3)
````

### Задача 1 (6 баллов)

Вам необходимо понять, какие слова встречаются в текстах на русском языке чаще всего. Программе на вход подаётся текст (одной строкой). Ваша задача:

- очистить текст от знаков препинания, пробелов, чисел и привести всё к нижнему регистру — *2 балла*;
- создать словарь, ключи которого — слова, а значения — то, сколько раз эти слова встречаются во введённом тексте — *4 балла*.

**Примечание.** Использование класса `Counter` из модуля `collections` (как и любых других модулей) <u>запрещено</u>.

**Пример:**

```
input:
Кроме того, я полагаю, что Карфаген должен быть разрушен.

output:
{'кроме': 1,
 'того': 1,
 'я': 1,
 'полагаю': 1,
 'что': 1,
 'карфаген': 1,
 'должен': 1,
 'быть': 1,
 'разрушен': 1}
```

**Критерии:**

См. в условии.

````python
# ваш код здесь
def word_counter(text):
    # очистка текста от знаков препинания, пробелов, чисел и приведение к нижнему регистру
    clean_text = ''.join(c.lower() for c in text if c.isalpha() or c.isspace())
    # разделение текста на слова и создание словаря
    words = clean_text.split()
    word_dict = {}
    for word in words:
        if word not in word_dict:
            word_dict[word] = 1
        else:
            word_dict[word] += 1
    return word_dict

text = 'Кроме того, я полагаю, что Карфаген должен быть разрушен.'
result = word_counter(text)
print(result)
````

### Задача 2 (10 баллов)

Вы участвуете в организации эксперимента и пишете программу, которая принимает на вход данные i-го объекта. Известно, что форматом хранения данных является таблица, где строка соответствует наблюдению (в данном случае — уникальному объекту), а столбец — переменной. Следовательно, в ячейке ij находится значение j-ой переменной (показателя) для i-го наблюдения (объекта). Напишите программу, которая будет принимать на вход по одной такой строке. Программа должна:

- принимать на вход разделитель (для элементов строки). Разделителем может быть `','`, `';'` и `' '` — *1 балл*;
- принимать на вход названия колонок (переменных) через указанный разделитель. Гарантируется, что разделитель всегда будет корректным — *1 балл*;
- принимать на вход типы колонок (переменных) через указанный разделитель; возможные типы — `int`, `float`, `bool`, `str` — *1 балл*;
- принимать на вход строки до тех пор, пока на вход не будет подано значение `'TRANSMISSION STOPPED'` — *2 балла*;
- если количество элементов в строке не соответствует заявленному ранее, печатать сообщение `'INVALID ELEMENTS NUMBER'` и начинать следующий ввод — *1 балл*;
- возвращать и печатать для пользователя словарь, в котором ключами будут названия, а значениями — списки со значениями каждой переменной для каждого объекта. Значения переменной должны быть приведены к соответствующему типу. Гарантируется, что входная спецификация типов корректна — *4 балла*.

**Примечание.** Использование модуля `pandas` <u>запрещено</u>.

**Пример:**

```
input:

Введите разделитель: ;
Введите названия колонок: height;weight;gender
Введите типы данных: float;float;str
190;84;male
160;50;female
175;65;male
183;100;male
TRANSMISSION STOPPED

output:

{'height': [190.0, 160.0, 175.0, 183.0],
 'weight': [84.0, 50.0, 65.0, 100.0],
 'gender': ['male', 'female', 'male', 'male']}
```

**Критерии:**

См. в условии.

````python
# ваш код здесь
def input_table():
    delimiter = input("Введите разделитель: ")
    columns = input("Введите названия колонок: ").split(delimiter)
    types = input("Введите типы данных: ").split(delimiter)
    
    # приводим названия колонок и типы данных к единому регистру
    columns = [col.lower() for col in columns]
    types = [typ.lower() for typ in types]
    
    # создаём словарь для хранения значений переменных
    table = {col: [] for col in columns}
    
    while True:
        row = input().split(delimiter)
        if row[0] == 'TRANSMISSION STOPPED':
            break
        
        # проверяем количество элементов в строке
        if len(row) != len(columns):
            print('INVALID ELEMENTS NUMBER')
            continue
        
        for i, value in enumerate(row):
            # приводим значение к соответствующему типу
            if types[i] == 'int':
                value = int(value)
            elif types[i] == 'float':
                value = float(value)
            elif types[i] == 'bool':
                value = bool(int(value))
            # если тип str, ничего не делаем
            # добавляем значение в словарь
            table[columns[i]].append(value)
    
    return table

input_table()
````
